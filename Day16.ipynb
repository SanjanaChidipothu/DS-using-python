## list comprehension

## syntax

[iterVariable for IterVar in groupOfElements]

```python
l1 = [i for i in range(1, 101)]

print(l1)
```

```python
l1 = [[i] for i in range(1, 101)]

print(l1)
```

```python
l1 = [i for i in range(1, 101) if i%2==0 or i%3==0]

print(l1)
```

```python
l1 = ["*" for i in range(1, 101)]

print(l1)
```

```python
l1 = [i for i in range(2000, 3000) if (i%400==0) or( i%100!=0 and i%4==0)]

print(l1)
```

```python
l1 = ["even" if i%2==0 else "odd" for i in range(1, 101)]

print(l1)
```

## Dictionary Comprehensions

create a dictionar key as the number and value as the square of the number of 1- 100

```python
d1 = {}

for num in range(1, 101):
    d1[num] = num**2
print(d1)
```

```python
d1 = {i : i**2 for i in range(1, 101)}

print(d1)
```

```python
d1 = {i : i + 5 for i in range(1, 101)}

print(d1)
```

```python
## create a dict with key as number and value as square of prev number

d1 = {i: (i-1)**2 for i in range(1, 101)}

print(d1)
```

```python
d1 = {i : i**2 for i in range(1, 101) if i%2==0}

print(d1)
```

```python
d1 = {i : "Even" if i%2==0 else "Odd" for i in range(1, 101) }

print(d1)
```

```python
d1 = {i : i**2 if i%2==0 for i in range(1, 101)}

print(d1)
```

create a dict such that {1:[1], 2:[1,2], 3:[1,2,3]}

```python
d1 = {i : [num for num in range(1,i+1)] for i in range(1, 11)}

print(d1)
```

```python
d1 = {i : ["*" for num in range(1,i+1)] for i in range(1, 11)}

print(d1)
```

```python
d1 = {i : ["*" for num in range(1,i+1)] for i in range(1, 101)}

d1
```

```python
l1 = [i for i in range(1, 101)]

d1 = {i : l1[: i+1] for i in range(1, 101)}

print(d1)
```

## Syntax of map

map(function : seqOfData)

It returns map object -> list/tuple/set

```python
l1 = list(map(int, input().split()))

print(l1)
```

```python
l1 = list(map(int, input().split()))

print(l1, sum(l1))
```

```python
l1 = map(lambda x: x[0], input().split())

print(list(l1))
```

```python
l1 = map(lambda x: int(x[0]), input().split())

print(list(l1))
```
