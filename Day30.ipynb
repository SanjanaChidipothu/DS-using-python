- kernel density estimator plot
- sns.kdeplot()
- data distributionbased on density

```python
sns.set_theme("poster")
sns.kdeplot(data=df1)
```

```python
sns.kdeplot(data=df1)
```

```python
sns.kdeplot(x = "tip",data=df1)
```

```python
sns.kdeplot(x="tip",y="total_bill",data=df1,hue="smoker")
```

```python
sns.kdeplot(x="tip",y="total_bill",data=df1,hue="sex")
```

```python
sns.kdeplot(x="tip",data=df1,bw_adjust=5)
```

```python
sns.kdeplot(x="tip",data=df1,bw_adjust=0.5)##less smoothing
```

## Joint plot

- sns.jointplot()
- draws a plot of two variables with bivariate and univariate graphs

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(data=df1)
```

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(df1["day"])
```

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(df1["day"],df1["sex"])
```

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(df1["total_bill"],df1["tip"],kind="kde")
```

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(df1["total_bill"],df1["tip"],kind="reg")##regression plot
```

```python
sns.set_theme(style = "darkgrid")
sns.jointplot(df1["total_bill"],df1["tip"],kind="hex")##hexagonal plot
```
