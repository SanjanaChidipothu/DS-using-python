## Concatenation of data frames

- pandas.concat((df1,df2))

```python
d1 = pd.DataFrame({"clg":["jntua","jntuv","jntuk","jntuh"], 
                   "temp":[35,45,44,43],
                   "humidity":[40,41,42,43]})
d1
```

```python
d2 = pd.DataFrame({"clg":["abc","def","ghi","jkl"], 
                   "temp":[35,45,44,43],
                   "humidity":[40,41,42,43]})
d2
```

```python
d3 = pd.concat((d1, d2))
d3
```

```python
d2 = pd.DataFrame({"clg":["abc","def","ghi","jkl"], 
                   "temp1":[35,45,44,43],
                   "humidity1":[40,41,42,43]})
d2
```

```python
d3 = pd.concat((d1, d2))
d3
```

```python
import pandas as pd
d3 = pd.concat((d1,d2), ignore_index= True )
d3
```

```python
import pandas as pd
d3 = pd.concat((d1,d2), keys = ["k1","k2"] )
d3
```

## Merging data frames

- df.merge()

```python
d1
```

```python
d2
```

```python
a = pd.DataFrame({"lkey":[11,22,33,44],"key":[33,44,55,66]})
b = pd.DataFrame({"rkey":[11,22,33,44],"key":[33,4,5,6]})

a
```

```python
b
```

```python
c = pd.merge(a,b) ##only common values are obtained
c
```

```python
c = pd.merge(a,b, indicator = True) ##only common values are obtained
c
```

```python
c = pd.merge(a,b,how= "left", indicator = True) ##only common values are obtained
c
```

```python
c = pd.merge(a,b,how= "right", indicator = True) ##only common values are obtained
c
```

```python
c = pd.merge(a,b,on = "key", indicator = True) ##only common values are obtained
c
```
