- Random forest classifier(more accuracy than any other algorithm) 
    - called as ensemble technique
    - Present in Ensemble classifier
    - Random forest - group of decision trees
    - We will consider the output of each decision tree and take the highest repeated values as our result
    


- If decision tree has overfitting(low bias, high variance) the use random forest 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
x = df.drop("target",axis =1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=25)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score
accuracy_score(y_test, y_pred)
```

```python
help(RandomForestClassifier())
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
## n_estimators no of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=200,stop=2000,num=10)]
## max_features is no of features at every split
max_features = ["auto","sqrt"]
##max_depth
max_depth = [int(x) for x in np.linspace(100,500,num=11)]

random_grid = {"n_estimators":n_estimators, "max_features":max_features,"max_depth":max_depth}
rfc = RandomizedSearchCV(estimator = model,param_distributions = random_grid,n_iter=100,cv=3,random_state=42,verbose=2,n_jobs=-1)
rfc.fit(x_train,y_train)
print(rfc.best)
```

```python
model = RandomForestClassifier(n_estimators=400, max_features= 'auto', max_depth= 180)
model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score
y_pred = model.predict(x_test)
cv_score = cross_val_score(model,x,y,cv=10)
print(accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_pred,y_test)
```
