```python
## no of lines ina file
## no of characters in a file
## no of words in a file


with open ("data2.txt", "r") as f1:
    k = f1.read()
    count = 0
    wcount = 0
    for i in k:
        if i == "\n":
            count = count + 1
            wcount = k.split()
print("no of lines in the file are", count)
print("no of words in the file are", len(wcount))
print("no of characters in the file are", len(k))
```

```python
with open("file1.txt", "r+") as f:
    f.write("Day12 Session") ## using with open technique the file automatically closes

print(f.read())
```

```python
## copying data from one file to many

with open("file1.txt", "r+") as f:
    data = f.read()

for i in range(1, 11):
    with open("file"+str(i)+".txt","r+") as f:
        f.write(data)
        
```
